```python
#find the batsman with max no of sixes
six=data[data['batsman_runs']==6]
```

```python
#find the batsman with most number of 4's and 6's in last 5 overs
data1=data[data['over']>15] 
data2=data1[(data1['batsman_runs']==6) | (data1['batsman_runs']==4)]
```

```python
data2.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(1).index[0]
```

```python
data3=data[data['batsman']=='V Kohli']
```

```python
data3.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
#create a function that can return the highest score of any batsman

def bat(batsman):
    data2=data[data['batsman']==batsman]
    return data2.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
bat('V Kohli')
```

```python
bat('RG Sharma')
```

```python
bat('CH Gayle')
```
